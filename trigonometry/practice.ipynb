## SKETCHING SINE AND COSINE

```python
# y = 3 sin(/2)
equation = lambda x: 3*np.sin(x/2)
x = np.linspace(-10, 10, 400)
y = equation(x)

label = "$3 \\cdot \\sin(\\frac{\\theta}{2})$"
points = [
  {"x": 0, "y": 0, "label": "$(0, 0)$"},
  {"x": math.pi, "y": 3, "label": "$(\\pi, 3)$"},
  {"x": 2*math.pi, "y": 0, "label": "$(2\\pi, 0)$"},
  {"x": 3*math.pi, "y": -3, "label": "$(3\\pi, -3)$"}
]
line = {
  "x": x,
  "y": y,
  "label": label,
  "points": points,
}

title = "$y = 3 \\cdot \\sin(\\frac{\\theta}{2})$"
lines = [line]
limits = {"x": (-5, 10), "y": (-5, 5)}

plot_lines(title, lines, limits=limits)
```

```python
# y = 2.6 cos(3)
equation = lambda x: 2.6 * np.cos(3 * x)
x = np.linspace(-2*math.pi, 2*math.pi, 400)
y = equation(x)

label = "$2.6 \\cdot \\cos(3\\theta)$"
points = [
  {"x": 0, "y": 2.6, "label": "$(0, 2.6)$"},
  {"x": math.pi/6, "y": 0, "label": "$(\\frac{\\pi}{6}, 0)$"},
  {"x": math.pi/3, "y": -2.6, "label": "$(\\frac{\\pi}{3}, -2.6)$"},
  {"x": math.pi/2, "y": 0, "label": "$(\\frac{\\pi}{2}, 0)$"},
]
line = {
  "x": x,
  "y": y,
  "label": label,
  "points": points,
}

title = "$y = 2.6 \\cdot \\cos(3\\theta)$"
lines = [line]
limits = {"x": (-3, 3), "y": (-3, 3)}

plot_lines(title, lines, limits=limits)
```

```python
# y = 4 cos(/3)
equation = lambda x: -4 * np.cos(x / 3)
x = np.linspace(-6*math.pi, 6*math.pi, 400)
y = equation(x)

label = "$-4 \\cdot \\cos(\\frac{\\theta}{3})$"
points = [
  {"x": -3*math.pi, "y": 4, "label": "$(-3\\pi, 4)$"},
  {"x": 0, "y": -4, "label": "$(0, -4)$"},
  {"x": 3*math.pi/2, "y": 0, "label": "$(\\frac{3\\pi}{2}, 0)$"},
  {"x": 3*math.pi, "y": 4, "label": "$(3\\pi, 4)$"},
]
line = {
  "x": x,
  "y": y,
  "label": label,
  "points": points,
}

title = "$y = -4 \\cdot \\cos(\\frac{\\theta}{3})$"
lines = [line]
limits = {"x": (-10, 10), "y": (-5, 5)}

plot_lines(title, lines, limits=limits)
```

```python
# y = 2 cos  and y = sin 2
equation1 = lambda x: 2 * np.cos(x)
equation2 = lambda x: np.sin(2 * x)
x = np.linspace(-2*math.pi, 2*math.pi, 400)

y1 = equation1(x)
points1 = [
  {"x": -math.pi, "y": -2, "label": "$(-\\pi, -2)$"},
  {"x": 0, "y": 2, "label": "$(0, 2)$"},
  {"x": math.pi, "y": -2, "label": "$(\\pi, -2)$"},
]
line1 = {
  "x": x,
  "y": y1,
  "label": "$2 \\cdot \\cos(\\theta)$",
  "points": points1,
}

y2 = equation2(x)
points2 = [
  {"x": -math.pi/4, "y": -1, "label": "$(-\\frac{\\pi}{4}, -1)$"},
  {"x": 0, "y": 0, "label": "$(0, 0)$"},
  {"x": math.pi/4, "y": 1, "label": "$(\\frac{\\pi}{4}, 1)$"},
]
line2 = {
  "x": x,
  "y": y2,
  "points": points2,
  "label": "$\\sin(2\\theta)$",
  "color": "red"
}

title = "$y = 2 \\cdot \\cos(\\theta)$ and $y = \\sin(2\\theta)$"
lines = [line1, line2]
limits = {"x": (-math.pi, math.pi), "y": (-3, 3)}

plot_lines(title, lines, limits=limits)
```

## SKETCHING COSECANT AND SECANT

```python
# y= csc(/2)
equation = lambda x: 1 / np.sin(x / 2)
x = np.linspace(-10, 10, 400)
y = equation(x)

label = "$\\csc(\\frac{\\theta}{2})$"

# Find peaks and segment lines
lines = []
for seg in get_segment_lines_by_peaks(x, y):
  lines.append(seg)

title = "$y = \\csc(\\frac{\\theta}{2})$"
limits = {"x": (-10, 10), "y": (-5, 5)}

plot_lines(title, lines, limits=limits)
```
