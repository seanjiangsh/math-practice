## Trigonometry

```python
# y = sin(x)
x = np.linspace(-10, 10, 400)
y1 = (lambda x: np.sin(x))(x)
label_1 = "$\\sin(x)$"
point_1 = [
    {"x": 0, "y": 0, "label": "$(0, 0)$"},
    {"x": math.pi/2, "y": 1, "label": "$(\\frac{\\pi}{2}, 1)$"},
    {"x": math.pi, "y": 0, "label": "$(\\pi, 0)$"},
    {"x": 3*math.pi/2, "y": -1, "label": "$(\\frac{3\\pi}{2}, -1)$"}
]
line_1 = {
    "x": x,
    "y": y1,
    "label": label_1,
    "points": point_1
}

# y = 3 * sin(x/2)
y2 = (lambda x: 3*np.sin(x/2))(x)
label_2 = "$3 \\cdot \\sin(\\frac{x}{2})$"
point_2 = [
    {"x": 0, "y": 0, "label": "$(0, 0)$"},
    {"x": math.pi, "y": 3, "label": "$(\\pi, 3)$"},
    {"x": 2*math.pi, "y": 0, "label": "$(2\\pi, 0)$"},
    {"x": 3*math.pi, "y": -3, "label": "$(3\\pi, -3)$"}
]
line_2 = {
    "x": x,
    "y": y2,
    "label": label_2,
    "points": point_2,
    "color": "red"
}

title = "$y = 3 \\cdot \\sin(\\frac{x}{2})$"
lines = [line_1, line_2]
limits = {"x": (-5, 10), "y": (-5, 5)}

plot_lines(title, lines, limits=limits)
```
