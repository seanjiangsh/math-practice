## Cosecant and Secant

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Define x values
x = np.linspace(-10, 10, 800)
y2 = 1 / np.sin(2 * x) + 1  # Compute y = 1 + csc(2x)

# Find local minima (valleys) and maxima (peaks)
peaks, _ = find_peaks(y2, height=0)  # Find peaks (top of U-shape)
valleys, _ = find_peaks(-y2, height=0)  # Find valleys (bottom of U-shape)

jumps = np.where(np.abs(np.diff(y2)) > 10)[0]  # Detect large jumps

print("Peaks:", peaks, len(peaks))
print("jumps:", jumps, len(jumps))

# Combine peak and valley indices as segment boundaries
segment_boundaries = np.sort(np.concatenate((peaks, valleys)))

# Split data at these boundaries
x_segments = np.split(x, segment_boundaries)
y_segments = np.split(y2, segment_boundaries)

# Plot the segmented curve
plt.figure(figsize=(8, 6))
for x_seg, y_seg in zip(x_segments, y_segments):
    plt.plot(x_seg, y_seg, 'r')  # Plot only continuous segments

# Formatting
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--')
plt.legend([r"$y = 1 + \csc(2x)$"])
plt.show()
```
