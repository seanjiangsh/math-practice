```python
# r^2 * sin(2θ) = 1
# we can express the equation as y = 1 / (2 * x)
title = "$r^2sin(2\\theta) = 1$"
x = np.linspace(-5, 5, 1000)
y = (lambda x: 1 / (2 * x))(x)

# Find peaks and segment lines
lines = []
for seg in get_segment_lines_by_peaks(x, y):
    lines.append(seg)
    
limits = {"x": (-3, 3), "y": (-3, 3)}

plot_lines(title, lines, limits=limits)
```

```python
# rcos(θ)  - tan(θ) + 2 = 0
# we can express the equation as y = x(x+2)
title = "$rcos(\\theta) - tan(\\theta) + 2 = 0$"
x = np.linspace(-5, 5, 1000)
y = (lambda x: x * (x + 2))(x)
lines = [{'x': x, 'y': y}]
limits = {"x": (-5, 3), "y": (-2, 6)}

plot_lines(title, lines, limits=limits)
```

```python
# r^2 = sin(θ) / cos(θ)
# we can express the equation as y = x^3
title = "$r^2 = \\frac{sin(\\theta)}{cos(\\theta)}$"
x = np.linspace(-5, 5, 1000)
y = (lambda x: x**3)(x)
lines = [{'x': x, 'y': y}]
limits = {"x": (-2, 2), "y": (-2, 2)}

plot_lines(title, lines, limits=limits)
```
