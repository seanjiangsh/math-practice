## Gauss-Jordan Elimination Examples

Testing the Gauss-Jordan elimination function to solve systems of linear equations.

```python
import sys
sys.path.append('../utils')
from preCalculus import gauss_jordan_elimination, solve_linear_system
import numpy as np
```

```python
# Example 1: System with unique solution
# 2x + 3y = 7
# 4x - y = 5

print("Example 1: System with unique solution")
print("2x + 3y = 7")
print("4x - y = 5")
print()

# Augmented matrix [A|b]
matrix1 = [
    [2, 3, 7],   # 2x + 3y = 7
    [4, -1, 5]   # 4x - y = 5
]

result1 = solve_linear_system(matrix1, ['x', 'y'])
print("RREF Matrix:")
print(result1['rref'])
print(f"\nStatus: {result1['status']}")
if result1['status'] == 'unique':
    print(f"Solution: {result1['solution']}")

# Verification
x, y = result1['solution']['x'], result1['solution']['y']
print(f"\nVerification:")
print(f"2({x}) + 3({y}) = {2*x + 3*y} (should be 7)")
print(f"4({x}) - 1({y}) = {4*x - y} (should be 5)")
```

```python
# Example 2: System with no solution (inconsistent)
# x + 2y = 3
# 2x + 4y = 10  (This is inconsistent because 2*(x + 2y) = 2*3 = 6, not 10)

print("Example 2: System with no solution")
print("x + 2y = 3")
print("2x + 4y = 10")
print()

matrix2 = [
    [1, 2, 3],    # x + 2y = 3
    [2, 4, 10]    # 2x + 4y = 10
]

result2 = solve_linear_system(matrix2, ['x', 'y'])
print("RREF Matrix:")
print(result2['rref'])
print(f"\nStatus: {result2['status']}")
print(f"Solution: {result2['solution']}")
```

```python
# Example 3: System with infinite solutions
# x + 2y = 4
# 2x + 4y = 8  (This is dependent: 2*(x + 2y) = 2*4 = 8)

print("Example 3: System with infinite solutions")
print("x + 2y = 4")
print("2x + 4y = 8")
print()

matrix3 = [
    [1, 2, 4],   # x + 2y = 4
    [2, 4, 8]    # 2x + 4y = 8
]

result3 = solve_linear_system(matrix3, ['x', 'y'])
print("RREF Matrix:")
print(result3['rref'])
print(f"\nStatus: {result3['status']}")
print(f"Free variables: {result3['free_variables']}")
print("Solution (in terms of free variables):")
for var, expr in result3['solution'].items():
    print(f"  {expr}")
```

```python
# Example 4: 3x3 system with unique solution
# x + 2y - z = 1
# 2x - y + 3z = 14
# 3x + y + z = 11

print("Example 4: 3x3 system with unique solution")
print("x + 2y - z = 1")
print("2x - y + 3z = 14")
print("3x + y + z = 11")
print()

matrix4 = [
    [1, 2, -1, 1],    # x + 2y - z = 1
    [2, -1, 3, 14],   # 2x - y + 3z = 14
    [3, 1, 1, 11]     # 3x + y + z = 11
]

result4 = solve_linear_system(matrix4, ['x', 'y', 'z'])
print("RREF Matrix:")
print(result4['rref'])
print(f"\nStatus: {result4['status']}")
if result4['status'] == 'unique':
    print(f"Solution: {result4['solution']}")
    
    # Verification
    x, y, z = result4['solution']['x'], result4['solution']['y'], result4['solution']['z']
    print(f"\nVerification:")
    print(f"({x}) + 2({y}) - ({z}) = {x + 2*y - z} (should be 1)")
    print(f"2({x}) - ({y}) + 3({z}) = {2*x - y + 3*z} (should be 14)")
    print(f"3({x}) + ({y}) + ({z}) = {3*x + y + z} (should be 11)")
```

```python
# Example 5: 3x3 system with infinite solutions
# x + y + z = 6
# 2x + y - z = 1
# 3x + 2y = 7  (Note: This is dependent on the first two equations)

print("Example 5: 3x3 system with infinite solutions")
print("x + y + z = 6")
print("2x + y - z = 1")
print("3x + 2y = 7")
print()

matrix5 = [
    [1, 1, 1, 6],    # x + y + z = 6
    [2, 1, -1, 1],   # 2x + y - z = 1
    [3, 2, 0, 7]     # 3x + 2y = 7
]

result5 = solve_linear_system(matrix5, ['x', 'y', 'z'])
print("RREF Matrix:")
print(result5['rref'])
print(f"\nStatus: {result5['status']}")
print(f"Free variables: {result5['free_variables']}")
if result5['solution']:
    print("Solution (in terms of free variables):")
    for var, expr in result5['solution'].items():
        print(f"  {expr}")
```

```python
# Example 6: Testing the raw Gauss-Jordan elimination function
print("Example 6: Testing raw Gauss-Jordan elimination")
print("Matrix from Example 1:")

matrix_test = np.array([
    [2, 3, 7],
    [4, -1, 5]
], dtype=float)

print("Original matrix:")
print(matrix_test)

rref, status = gauss_jordan_elimination(matrix_test)
print(f"\nRREF matrix:")
print(rref)
print(f"Status: {status}")

# Show step-by-step RREF process for educational purposes
print("\n" + "="*50)
print("Manual verification of RREF steps:")
print("="*50)

# Step 1: Make first element 1
A = matrix_test.copy()
print(f"Original: \n{A}")

# R1 = R1 / 2
A[0] = A[0] / 2
print(f"R1 = R1/2: \n{A}")

# Step 2: Eliminate below pivot
# R2 = R2 - 4*R1
A[1] = A[1] - 4 * A[0]
print(f"R2 = R2 - 4*R1: \n{A}")

# Step 3: Make second pivot 1
# R2 = R2 / (-7)
A[1] = A[1] / A[1, 1]
print(f"R2 = R2/{A[1,1]}: \n{A}")

# Step 4: Eliminate above second pivot
# R1 = R1 - (3/2)*R2
A[0] = A[0] - A[0, 1] * A[1]
print(f"R1 = R1 - {A[0,1]}*R2: \n{A}")

print(f"\nFinal RREF matches our function: {np.allclose(A, rref)}")
```

## Matrix Transformations

```python
# Original triangle points
triangle_points = np.array([
  [2, 3],    # Point 1: (2, 3)
  [-3, -1],  # Point 2: (-3, -1)
  [1, -4]    # Point 3: (1, -4)
])

# Transpose to get points in columns for matrix multiplication
triangle_matrix = triangle_points.T

# Transformation matrix
trans_matrix = np.array([
  [1, 2],
  [2, -1]
])

print("Original triangle matrix (points as columns):")
print(triangle_matrix)
print("Shape of triangle matrix:", triangle_matrix.shape)
print("Shape of transformation matrix:", trans_matrix.shape)

# Apply transformation: multiply transformation matrix by triangle_matrix
translated_triangle_matrix = np.dot(trans_matrix, triangle_matrix)
print("\nTransformed triangle matrix:")
print(translated_triangle_matrix)

# Convert back to points format (transpose back)
transformed_points = translated_triangle_matrix.T
print("\nTransformed points:")
print(transformed_points)

# Create polygons for visualization using the refactored format
original_triangle = {
    'coordinates': triangle_points,
    'point_labels': ['P1(2,3)', 'P2(-3,-1)', 'P3(1,-4)'],
    'color': 'blue',
    'label': 'Original Triangle'
}

transformed_triangle = {
    'coordinates': transformed_points,
    'point_labels': ["P1'(8,1)", "P2'(-5,-5)", "P3'(-7,6)"],
    'color': 'red',
    'label': 'Transformed Triangle'
}

title = 'Triangle Transformation: Original vs Transformed'
limits = {'x': (-10, 10), 'y': (-10, 10)}

plot_geometry(title, polygons=[original_triangle, transformed_triangle], limits=limits)
```

```python
square_points = np.array([
  [0, 0],
  [0, 5],
  [-5, 5],
  [-5, 0]
])
square_matrix = square_points.T

trans_matrix = np.array([
  [1, -1],
  [0, 2]
])

translated_square = np.dot(trans_matrix, square_matrix)
print("\nTransformed square matrix:")
print(translated_square)

# Format point labels using lambda
format_label = lambda i, pt: f"P{i+1}({pt[0]},{pt[1]})"
original_square = {
  'coordinates': square_points,
  'point_labels': [format_label(i, pt) for i, pt in enumerate(square_points)],
  'color': 'blue',
  'label': 'Original Square'
}
transformed_square = {
  'coordinates': translated_square.T,
  'point_labels': [format_label(i, pt) for i, pt in enumerate(translated_square.T)],
  'color': 'red',
  'label': 'Transformed Square'
}

title = "Transformation of a Square"
limits = {
  'x': (-15, 5),
  'y': (-5, 15)
}

plot_geometry(title, polygons=[original_square, transformed_square], limits=limits)
```

```python
hexagon_points = np.array([
  [5, 0],
  [3, 2],
  [-2, 3],
  [-7, 0],
  [-5, -5],
  [4, -4]
])
hexagon_matrix = hexagon_points.T

trans_matrix = np.array([
  [0, 2],
  [1, 1]
])

translated_hexagon = np.dot(trans_matrix, hexagon_matrix)
print("\nTransformed hexagon matrix:")
print(translated_hexagon)

format_label = lambda i, pt: f"P{i+1}({pt[0]},{pt[1]})"
original_hexagon = {
  'coordinates': hexagon_points,
  'point_labels': [format_label(i, pt) for i, pt in enumerate(hexagon_points)],
  'color': 'blue',
  'label': 'Original Hexagon'
}
transformed_hexagon = {
  'coordinates': translated_hexagon.T,
  'point_labels': [format_label(i, pt) for i, pt in enumerate(translated_hexagon.T)],
  'color': 'red',
  'label': 'Transformed Hexagon'
}

title = "Transformation of a Hexagon"
limits = {
  'x': (-12, 8),
  'y': (-12, 8)
}

plot_geometry(title, polygons=[original_hexagon, transformed_hexagon], limits=limits)
```
