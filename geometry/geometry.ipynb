## Plotting Lines

$\overline{AB} \perp \overline{CD}$ and they intersect at E. 

Draw a sketch of this and include all necessary labels on your diagram.

```python
ab_x = np.array([-5, 5])
ab_y = np.array([0, 0])
ab_points = ['A', 'B']
ab = {'x': ab_x, 'y': ab_y, 'points': ab_points}

cd_x = np.array([0, 0, 0])
cd_y = np.array([5, 0, -5])
cd_points = ['C', 'E', 'D']
cd = {'x': cd_x, 'y': cd_y, 'points': cd_points}

title = '$\\overline{AB} \\perp \\overline{CD}$ and they intersect at E'
plot_geometry(title, [ab, cd])
```

Sketch the following: $\overline{AB}$ lies on plane DEF and C is contained in $\overline{AB}$

```python
abc_x = np.array([-3, 1, 3])
abc_y = np.array([1, 1, 1])
abc_points = ['A', 'C', 'B']
abc = {'x': abc_x, 'y': abc_y, 'points': abc_points}

polygon_x = np.array([-5, 5, 7, -7])
polygon_y = np.array([5, 5, -3, -3])
polygon_points = ['D', 'E', 'F', 'G']
polygon = {
    'x': polygon_x,
    'y': polygon_y,
    'points': polygon_points
}

title = '$\\overline{AB}$ lies on plane DEF and C is contained in $\\overline{AB}$'
plot_geometry(title, [abc], [polygon])
```

## Plotting Geometry Shapes

Graph the parallelogram JKLM, if J(0,0), K(1,3), L(−5,3), and

```python
polygon_x = np.array([0, 1, -5, -6])
polygon_y = np.array([0, 3, 3, 0])
polygon_points = ['J', 'K', 'M', 'L']
polygon = {
    'x': polygon_x,
    'y': polygon_y,
    'points': polygon_points
}

title = 'Parallelogram JKLM'

plot_geometry(title, polygons=[polygon])
```

$\triangle{SAM}$ is a 30-60-90 triangle:

* S (0,0)
* A (4\sqrt{3}, 4)
* M (4\sqrt{3}, 0)

Which angle is the $60\degree$ angle?

```python
triangle_x = np.array([0, 4*np.sqrt(3), 4*np.sqrt(3)])
triangle_y = np.array([0, 4, 0])
triangle_points = ['S', 'A', 'M']
triangle = {
    'x': triangle_x,
    'y': triangle_y,
    'points': triangle_points
}

title = 'Triangle SAM'

plot_geometry(title, polygons=[triangle])
```

A rectangle has a diagonal with endpoints at (5,1) and (14,7). Find the 
area of this rectangle under the translation (x, y) →(x−5, y−4).

```python
# rectangle has a diagonal with endpoints at (5,1) and (14,7)
rect_width = 14 - 5
rect_height = 7 - 1
print(f'Width of rectangle: {rect_width}')
print(f'Height of rectangle: {rect_height}')

rectangle_x = np.array([5, 5, 14, 14])
rectangle_y = np.array([1, 7, 7, 1])
rectangle_points = ['A', 'B', 'C', 'D']
rectangle = {
    'x': rectangle_x,
    'y': rectangle_y,
    'points': rectangle_points
}

translated_rectangle_x = rectangle_x - 5
translated_rectangle_y = rectangle_y - 4
translated_rectangle_points = ["A'", "B'", "C'", "D'"]
translated_rectangle = {
    'x': translated_rectangle_x,
    'y': translated_rectangle_y,
    'points': translated_rectangle_points,
    'color': 'purple'
}

title = 'Rectangle ABCD and its translation A\'B\'C\'D\''
limits={'x': [-5, 15], 'y': [-10, 10]}

plot_geometry(title, polygons=[rectangle, translated_rectangle], limits=limits)
```

$\triangle{ABC}=(4,1), (1,5), (1,1)$, graph $\triangle{ABC}$ under a rotation $90\degree$ counterclockwise

```python
# triangle ABC = (4,1), (1,5), (1,1)
# triangle A'B'C' = (-1,4), (-5,1), (-1,1)

triangle_abc_x = np.array([4, 1, 1])
triangle_abc_y = np.array([1, 5, 1])
triangle_abc_points = ['A', 'B', 'C']
triangle_abc = {
    'x': triangle_abc_x,
    'y': triangle_abc_y,
    'points': triangle_abc_points
}

triangle_abc_rotated_x = np.array([-1, -5, -1])
triangle_abc_rotated_y = np.array([4, 1, 1])
triangle_abc_rotated_points = ["A'", "B'", "C'"]
triangle_abc_rotated = {
    'x': triangle_abc_rotated_x,
    'y': triangle_abc_rotated_y,
    'points': triangle_abc_rotated_points,
    'color': 'purple'
}

title = 'Triangle ABC and its rotation A\'B\'C\''

plot_geometry(title, polygons=[triangle_abc, triangle_abc_rotated])
```
