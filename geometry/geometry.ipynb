## Plotting Lines

$\overline{AB} \perp \overline{CD}$ and they intersect at E. 

Draw a sketch of this and include all necessary labels on your diagram.

```python
# Line AB: horizontal line from (-5, 0) to (5, 0)
ab_coordinates = np.array([[-5, 0], [5, 0]])
ab_point_labels = ['A', 'B']
ab = {'coordinates': ab_coordinates, 'point_labels': ab_point_labels}

# Line CD: vertical line from (0, 5) through E(0, 0) to (0, -5)
cd_coordinates = np.array([[0, 5], [0, 0], [0, -5]])
cd_point_labels = ['C', 'E', 'D']
cd = {'coordinates': cd_coordinates, 'point_labels': cd_point_labels}

title = '$\\overline{AB} \\perp \\overline{CD}$ and they intersect at E'
plot_geometry(title, [ab, cd])
```

Sketch the following: $\overline{AB}$ lies on plane DEF and C is contained in $\overline{AB}$

```python
# Line ABC: horizontal line from A(-3, 1) through C(1, 1) to B(3, 1)
abc_coordinates = np.array([[-3, 1], [1, 1], [3, 1]])
abc_point_labels = ['A', 'C', 'B']
abc = {'coordinates': abc_coordinates, 'point_labels': abc_point_labels}

# Plane DEFG: quadrilateral
polygon_coordinates = np.array([[-5, 5], [5, 5], [7, -3], [-7, -3]])
polygon_point_labels = ['D', 'E', 'F', 'G']
polygon = {
    'coordinates': polygon_coordinates,
    'point_labels': polygon_point_labels
}

title = '$\\overline{AB}$ lies on plane DEF and C is contained in $\\overline{AB}$'
plot_geometry(title, [abc], [polygon])
```

## Plotting Geometry Shapes

Graph the parallelogram JKLM, if J(0,0), K(1,3), L(5,3), and 
M(6,0)

```python
# Parallelogram JKLM with vertices at J(0,0), K(1,3), L(-5,3), M(-6,0)
polygon_coordinates = np.array([[0, 0], [1, 3], [-5, 3], [-6, 0]])
polygon_point_labels = ['J', 'K', 'L', 'M']
polygon = {
    'coordinates': polygon_coordinates,
    'point_labels': polygon_point_labels
}

title = 'Parallelogram JKLM'

plot_geometry(title, polygons=[polygon])
```

$\triangle{SAM}$ is a 30-60-90 triangle:

* S (0,0)
* A (4\sqrt{3}, 4)
* M (4\sqrt{3}, 0)

Which angle is the $60\degree$ angle?

```python
# Triangle SAM: S(0,0), A(43, 4), M(43, 0)
triangle_coordinates = np.array([[0, 0], [4*np.sqrt(3), 4], [4*np.sqrt(3), 0]])
triangle_point_labels = ['S', 'A', 'M']
triangle = {
    'coordinates': triangle_coordinates,
    'point_labels': triangle_point_labels
}

title = 'Triangle SAM'

plot_geometry(title, polygons=[triangle])
```

A rectangle has a diagonal with endpoints at (5,1) and (14,7). Find the 
area of this rectangle under the translation (x, y) (x5, y4).

```python
# rectangle has a diagonal with endpoints at (5,1) and (14,7)
rect_width = 14 - 5
rect_height = 7 - 1
print(f'Width of rectangle: {rect_width}')
print(f'Height of rectangle: {rect_height}')

# Rectangle ABCD vertices: A(5,1), B(5,7), C(14,7), D(14,1)
rectangle_coordinates = np.array([[5, 1], [5, 7], [14, 7], [14, 1]])
rectangle_point_labels = ['A', 'B', 'C', 'D']
rectangle = {
    'coordinates': rectangle_coordinates,
    'point_labels': rectangle_point_labels
}

# Apply translation (x, y)  (x-5, y-4)
translated_coordinates = rectangle_coordinates - np.array([5, 4])
translated_point_labels = ["A'", "B'", "C'", "D'"]
translated_rectangle = {
    'coordinates': translated_coordinates,
    'point_labels': translated_point_labels,
    'color': 'purple'
}

title = 'Rectangle ABCD and its translation A\'B\'C\'D\''
limits={'x': [-5, 15], 'y': [-10, 10]}

plot_geometry(title, polygons=[rectangle, translated_rectangle], limits=limits)
```

$\triangle{ABC}=(4,1), (1,5), (1,1)$, graph $\triangle{ABC}$ under a rotation $90\degree$ counterclockwise

```python
# Triangle ABC with vertices at (4,1), (1,5), (1,1)
triangle_abc_coordinates = np.array([[4, 1], [1, 5], [1, 1]])
triangle_abc_point_labels = ['A', 'B', 'C']
triangle_abc = {
    'coordinates': triangle_abc_coordinates,
    'point_labels': triangle_abc_point_labels
}

# Triangle A'B'C' after 90 counterclockwise rotation: (-1,4), (-5,1), (-1,1)
triangle_abc_rotated_coordinates = np.array([[-1, 4], [-5, 1], [-1, 1]])
triangle_abc_rotated_point_labels = ["A'", "B'", "C'"]
triangle_abc_rotated = {
    'coordinates': triangle_abc_rotated_coordinates,
    'point_labels': triangle_abc_rotated_point_labels,
    'color': 'purple'
}

title = 'Triangle ABC and its rotation A\'B\'C\''

plot_geometry(title, polygons=[triangle_abc, triangle_abc_rotated])
```

$\triangle{ABC}=(-3,6), (2,3), (-4,-5)$ undergoes the translation described by $T(x,y) = (x+2,y3)$, what will be the the points of $\triangle{A'B'C'}$? Draw $\triangle{A'B'C'}$ on the coordinate plane.

```python
# Triangle ABC with vertices at (-3,6), (2,3), (-4,-5)
triangle_abc_coordinates = np.array([[-3, 6], [2, 3], [-4, -5]])
triangle_abc_point_labels = ['A', 'B', 'C']
triangle_abc = {
    'coordinates': triangle_abc_coordinates,
    'point_labels': triangle_abc_point_labels
}

# Triangle A'B'C' after translation T(x,y) = (x+2, y-3): (-1,3), (4,0), (-2,-8)
triangle_abc_translated_coordinates = np.array([[-1, 3], [4, 0], [-2, -8]])
triangle_abc_translated_point_labels = ["A'", "B'", "C'"]
triangle_abc_translated = {
    'coordinates': triangle_abc_translated_coordinates,
    'point_labels': triangle_abc_translated_point_labels,
    'color': 'purple'
}

title = 'Triangle ABC and its translation A\'B\'C\''

plot_geometry(title, polygons=[triangle_abc, triangle_abc_translated])
```
