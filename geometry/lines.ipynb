## Plotting Lines

$\overline{AB} \perp \overline{CD}$ and they intersect at E. 

Draw a sketch of this and include all necessary labels on your diagram.

```python
plt.figure(figsize=(6, 6))
setup_plot()

point_label_offset = 0.3  # Adjust this value to shift horizontally

# Plot the line AB
ab_x_values = np.array([-5, 5])
ab_y_values = np.array([0, 0])
plt.plot(ab_x_values, ab_y_values, 'o-', markersize=5, color='blue')

# Label the line points
point_labels = ['A', 'B']
for i, txt in enumerate(point_labels):
    plt.text(
        ab_x_values[i] + point_label_offset,
        ab_y_values[i] + point_label_offset,
        txt,
        fontsize=12
    )

# Plot the line CD
cd_x_values = np.array([0, 0, 0])
cd_y_values = np.array([5, 0, -5])
plt.plot(cd_x_values, cd_y_values, 'o-', markersize=5, color='blue')

# Label the line points
point_labels = ['C', 'E', 'D']
for i, txt in enumerate(point_labels):
    plt.text(
        cd_x_values[i] + point_label_offset,
        cd_y_values[i] + point_label_offset,
        txt,
        fontsize=12
    )

# Add grid and labels
title = '$\\overline{AB} \\perp \\overline{CD}$ and they intersect at E'
plt.title(title)
plt.xlabel('x')
plt.ylabel('y', rotation=0)

# Show plot
plt.show()
```

Sketch the following: $\overline{AB}$ lies on plane DEF and C is contained in $\overline{AB}$

```python
plt.figure(figsize=(6, 6))
setup_plot()

point_label_offset = 0.3  # Adjust this value to shift horizontally

# Plot the line ABC
line_x_values = np.array([-3, 1, 3])
line_y_values = np.array([1, 1, 1])
plt.plot(line_x_values, line_y_values, 'o-', markersize=5, color='blue')

# Label the line points
point_labels = ['A', 'C', 'B']
for i, txt in enumerate(point_labels):
    plt.text(
        line_x_values[i] + point_label_offset,
        line_y_values[i] + point_label_offset,
        txt,
        fontsize=12
    )

# Define the x and y coordinates for the polygon vertices
x_values = np.array([-5, 5, 7, -7, -5])
y_values = np.array([5, 5, -3, -3, 5])

# Use plt.fill to fill the polygon defined by the points
plt.fill(x_values, y_values, color="lightblue", alpha=0.5)

# Plot the polygon edges
plt.plot(x_values, y_values, 'o-', markersize=5, color='blue')

# Label the polygon points
point_labels = ['D', 'E', 'F', 'G']  # Labels for the points

for i, txt in enumerate(point_labels):
    plt.text(
        x_values[i] + point_label_offset,
        y_values[i] + point_label_offset,
        txt,
        fontsize=12
    )
    
# Add grid and labels
title = '$\\overline{AB}$ lies on plane DEF and C is contained in $\\overline{AB}$'
plt.title(title)
plt.xlabel('x')
plt.ylabel('y', rotation=0)

# Show plot
plt.show()
```
