## General Log Rule (Inverse Functions)

### For $log_a{x} = y$

equals:
- $a^y = x$

inverse:
- $a^x = y$
- $\log_a{y} = x$

```python
from IPython.display import display, Math, Markdown

display(Markdown('The functions we are plotting are:'))
display(Math('$y = 2^x$'))
display(Math('$x = 2^y$'))
display(Markdown('The inverse line is the reflection of the original line over the line $y = x$.'))

# y = 2^x
# x = 2^y

x1 = np.linspace(-10, 10, 100)
y1 = 2**x1

y2 = np.linspace(-10, 10, 100)
x2 = 2**y2

# the inverse line is the reflection of the original line over the line y = x
x3 = np.linspace(-10, 10, 100)
y3 = x3

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = 2^x$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$x = 2^y$',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': '$y = x$',
    'color': 'orange',
    'linestyle': '--'
}

lines = [line_1, line_2, line_3]
title = 'Example of General Log Rule (Inverse Functions)'

plot_lines(title, lines)
```

## Graphing transformations of log functions

```python
# y = -ln(-x+3)
# 1. y = ln(x) (parent function)
# 2. y = ln(x+3) (horizontal shift to the right by 3)
# 3. y = ln(-x+3) (horizontal reflection over y-axis)
# 4. y = -ln(-x+3) (vertical reflection over x-axis, final graph)

# 1. y = ln(x) (x = e^y)
y1 = np.linspace(-10, 10, 100)
x1 = np.exp(y1)
label1 = '$x = e^y$ (parent function)'

# 2. y = ln(x+3) (x = e^y + 3)
y2 = np.linspace(-10, 10, 100)
x2 = np.exp(y2) - 3
label2 = '$x = e^y - 3$ (horizontal shift to the right by 3)'

# 3. y = ln(-x+3) (x = -e^y + 3)
y3 = np.linspace(-10, 10, 100)
x3 = -np.exp(y3) + 3
label3 = '$x = -e^y + 3$ (horizontal reflection over y-axis)'

# 4. y = -ln(-x+3) (x = -e^-y + 3)
y4 = np.linspace(-10, 10, 100)
x4 = -np.exp(-y4) + 3
label4 = '$x = -e^{-y} + 3$ (vertical reflection over x-axis, final graph)'

line_1 = {
    'x': x1,
    'y': y1,
    'label': label1,
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': label2,
    'color': 'green',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': label3,
    'color': 'orange',
}
line_4 = {
    'x': x4,
    'y': y4,
    'label': label4,
    'color': 'purple',
}

lines = [line_1, line_2, line_3, line_4]
title = 'Transformations of $y = -\\ln(-x+3)$'

plot_lines(title, lines)
```

```python
# y = 2 * log_2(x-2)
# 1. y = log_2(x) (parent function)
# 2. y = log_2(x-2) (horizontal shift to the right by 2)
# 3. y = 2 * log_2(x-2) (vertical stretch by a factor of 2, final graph)

# 1. y = log_2(x) (x = 2^y)
y1 = np.linspace(-10, 10, 100)
x1 = 2**y1

# 2. y = log_2(x-2) (x = 2^y + 2)
y2 = np.linspace(-10, 10, 100)
x2 = 2**y2 + 2

# 3. y = 2 * log_2(x-2) (x = 2^(y/2) + 2)
y3 = np.linspace(-10, 10, 100)
x3 = 2**(y3/2) + 2

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$x = 2^y$ (parent function)',
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$x = 2^y + 2$ (horizontal shift to the right by 2)',
    'color': 'green',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': '$x = 2^{\\frac{y}{2}} + 2$ (vertical stretch by a factor of 2, final graph)',
    'color': 'purple',
}

lines = [line_1, line_2, line_3]
title = 'Transformations of $y = 2 \\cdot \\log_2(x-2)$'

plot_lines(title, lines)
```

```python
# y = -3 * log_7(x-2) + 4
# 1. y = log_7(x) (parent function)
# 2. y = log_7(x-2) (horizontal shift to the right by 2)
# 3. y = 3 * log_7(x-2) (vertical stretch by a factor of 3)
# 4. y = -3 * log_7(x-2) (vertical reflection over the x-axis)
# 5. y = -3 * log_7(x-2) + 4 (vertical shift 4 units up, final graph)

# 1. y = log_7(x) (x = 7^y)
y1 = np.linspace(-10, 10, 100)
x1 = 7**y1

# 2. y = log_7(x-2) (x = 7^y + 2)
y2 = np.linspace(-10, 10, 100)
x2 = 7**y2 + 2

# 3. y = 3 * log_7(x-2) (x = 7^(y/3) + 2)
y3 = np.linspace(-10, 10, 100)
x3 = 7**(y3/3) + 2

# 4. y = -3 * log_7(x-2) (x = 7^(-y/3) + 2)
y4 = np.linspace(-10, 10, 100)
x4 = 7**(-y4/3) + 2

# 5. y = -3 * log_7(x-2) + 4 (x = 7^((-y-4)/3) + 2)
y5 = np.linspace(-10, 10, 100)
x5 = 7**(-(y5-4)/3) + 2

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$x = 7^y$ (parent function)',
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$x = 7^y + 2$ (horizontal shift to the right by 2)',
    'color': 'green',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': '$x = 7^{\\frac{y}{3}} + 2$ (vertical stretch by a factor of 3)',
    'color': 'orange',
}
line_4 = {
    'x': x4,
    'y': y4,
    'label': '$x = 7^{-\\frac{y}{3}} + 2$ (vertical reflection over the x-axis)',
    'color': 'purple',
}
line_5 = {
    'x': x5,
    'y': y5,
    'label': '$x = 7^{-\\frac{y-4}{3}} + 2$ (vertical shift 4 units up, final graph)',
    'color': 'red',
}

lines = [line_1, line_2, line_3, line_4, line_5]
title = 'Transformations of $y = -3 \\cdot \\log_7(x-2) + 4$'

plot_lines(title, lines)
```
