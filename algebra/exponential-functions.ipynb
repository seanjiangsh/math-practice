## Transformations of exponential functions

```python
# y = 3^(x-2) - 3
# 1. y = 3^x (parent function)
# 2. y = 3^(x-2) (vertical shift 2 units down)
# 3. y = 3^(x-2) - 3 (vertical shift 3 units down, final graph)

x1 = np.linspace(-10, 10, 100)
y1 = 3**x1

x2 = np.linspace(-10, 10, 100)
y2 = 3**(x2-2)

x3 = np.linspace(-10, 10, 100)
y3 = 3**(x3-2) - 3

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = 3^x$ (parent function)',
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y = 3^{x-2}$ (vertical shift 2 units down)',
    'color': 'orange',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': '$y = 3^{x-2} - 3$ (final graph)',
    'color': 'green',
}

lines = [line_1, line_2, line_3]
title = 'Transformations of Exponential Functions'

plot_lines(title, lines)
```

```python
# y = -2 * 2^(x+3) + 1
# 1. y = 2^x (parent function)
# 2. y = 2^(x+3) (horizontal shift 3 units to the left)
# 3. y = 2 * 2^(x+3) (vertical stretch by a factor of 2)
# 4. y = -2 * 2^(x+3) (vertical reflection over the x-axis)
# 5. y = -2 * 2^(x+3) + 1 (vertical shift 1 unit up, final graph)

x1 = np.linspace(-10, 10, 100)
y1 = 2**x1

x2 = np.linspace(-10, 10, 100)
y2 = 2**(x2+3)

x3 = np.linspace(-10, 10, 100)
y3 = 2 * 2**(x3+3)

x4 = np.linspace(-10, 10, 100)
y4 = -2 * 2**(x4+3)

x5 = np.linspace(-10, 10, 100)
y5 = -2 * 2**(x5+3) + 1


line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = 2^x$ (parent function)',
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y = 2^{x+3}$ (horizontal shift 3 units to the left)',
    'color': 'orange',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': '$y = 2 * 2^{x+3}$ (vertical stretch by a factor of 2)',
    'color': 'green',
}
line_4 = {
    'x': x4,
    'y': y4,
    'label': '$y = -2 * 2^{x+3}$ (vertical reflection over the x-axis)',
    'color': 'red',
}
line_5 = {
    'x': x5,
    'y': y5,
    'label': '$y = -2 * 2^{x+3} + 1$ (vertical shift 1 unit up, final graph)',
    'color': 'purple',
}

lines = [line_1, line_2, line_3, line_4, line_5]
title = 'Transformations of Exponential Functions'

plot_lines(title, lines)

```

```python
# f(x) = (1/2)^(-x-2) - 5
# 1. f(x) = 1/2^x (parent function)
# 2. f(x) = 1/2^(x-2) (horizontal shift 2 units to the right)
# 3. f(x) = 1/2^(-x-2) (horizontal reflection over the y-axis)
# 4. f(x) = 1/2^(-x-2) - 5 (vertical shift 5 units down, final graph)

x1 = np.linspace(-10, 10, 100)
y1 = 1/2**x1

x2 = np.linspace(-10, 10, 100)
y2 = 1/2**(x2-2)

x3 = np.linspace(-10, 10, 100)
y3 = 1/2**(-x3-2)

x4 = np.linspace(-10, 10, 100)
y4 = 1/2**(-x4-2) - 5

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$f(x) = \\frac{1}{2}^x$ (parent function)',
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$f(x) = \\frac{1}{2}^{x-2}$ (horizontal shift 2 units to the right)',
    'color': 'orange',
}
line_3 = {
    'x': x3,
    'y': y3,
    'label': '$f(x) = \\frac{1}{2}^{-x-2}$ (horizontal reflection over the y-axis)',
    'color': 'green',
}
line_4 = {
    'x': x4,
    'y': y4,
    'label': '$f(x) = \\frac{1}{2}^{-x-2} - 5$ (vertical shift 5 units down, final graph)',
    'color': 'red',
}

lines = [line_1, line_2, line_3, line_4]
title = 'Transformations of Exponential Functions'

plot_lines(title, lines)
```
