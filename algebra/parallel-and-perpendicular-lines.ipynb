## Practices

```python
# y = -5x +2, y= 1/5x - 4
x1, y1 = get_slope_intercept_points(-5, 2).values()
x2, y2 = get_slope_intercept_points(1/5, -4).values()

# notice that the y-intercept is different for the second line
# it will change the intersection point

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = -5x + 2$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y = \\frac{1}{5}x - 4$',
    'color': 'orange'
}
title = 'Perpendicular Lines'

plot_lines(title, [line_1, line_2])
```

```python
# y = -1/3x + 2, y = -1/3x - 2
x1, y1 = get_slope_intercept_points(-1/3, 2).values()
x2, y2 = get_slope_intercept_points(-1/3, -2).values()
line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = -\\frac{1}{3}x + 2$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y = -\\frac{1}{3}x - 2$',
    'color': 'orange'
}
title = 'Parallel Lines'

plot_lines(title, [line_1, line_2])
```

```python
# y = -3x + 2
# y = 1/3x + 5
x1, y1 = get_slope_intercept_points(-3, 2).values()
x2, y2 = get_slope_intercept_points(1/3, 5).values()
line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = -3x + 2$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y = \\frac{1}{3}x + 5$',
    'color': 'orange'
}
title = 'Perpendicular Lines'

plot_lines(title, [line_1, line_2])
```

```python
# y = 2/3x
# y = 2/3x + 3
x1, y1 = get_slope_intercept_points(2/3, 0).values()
x2, y2 = get_slope_intercept_points(2/3, 3).values()
line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y = \\frac{2}{3}x$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y = \\frac{2}{3}x + 3$',
    'color': 'orange'
}
title = 'Parallel Lines'

plot_lines(title, [line_1, line_2])
```
