## Testing and Practicing

```python
# y^2 = x (not a function)

# Generate x values
x = np.linspace(0, 10, 100)  # Start from 0 to avoid complex numbers

# Calculate y values based on the equation y^2 = x
y_positive = np.sqrt(x)
y_negative = -np.sqrt(x)

lines = [
    {'x': x, 'y': y_positive, 'color': 'blue'},
    {'x': x, 'y': y_negative, 'color': 'red'}
]
title = '$y^2 = x$'

plot_lines(title, lines)
```

```python
# y = 1/6x
points = get_slope_intercept_points(1/6, 0)
title = '$y = \\frac{1}{6}x$'

plot_lines(title, [points])
```

```python
# f(x) = 6 / x
x1 = np.linspace(-10, -0.1, 100)
x2 = np.linspace(0.1, 10, 100)

# Calculate y values based on the equation f(x) = coefficient / x
y1 = 6 / x1
y2 = 6 / x2

lines = [
    {'x': x1, 'y': y1, 'color': 'blue'},
    {'x': x2, 'y': y2, 'color': 'red'}
]
title = '$f(x) = \\frac{6}{x}$'

plot_lines(title, lines)
```

```python
# f(x) = x^5 - 3x^3

x = np.linspace(-3, 3, 100)
y = (lambda x: x**5 - 3*x**3)(x)

lines = [{'x': x, 'y': y}]
title = '$f(x) = x^5 - 3x^3$'

plot_lines(title, lines)
```

```python
# f(x) = 5x^2 - x^4

x = np.linspace(-3, 3, 100)
y = (lambda x: 5*x**2 - x**4)(x)

lines = [{'x': x, 'y': y}]
title = '$f(x) = 5x^2 - x^4$'

plot_lines(title, lines)
```

```python
# f(x) = 5x^2 - x^3

x = np.linspace(-3, 3, 100)
y = (lambda x: 5*x**2 - x**3)(x)

lines = [{'x': x, 'y': y}]
title = '$f(x) = 5x^2 - x^3$'

plot_lines(title, lines)
```

```python
# f(x) = -x^4 - 6x^2

x = np.linspace(-3, 3, 100)
y = (lambda x: -1*x**4 - 6*x**2)(x)

lines = [{'x': x, 'y': y}]
title = '$f(x) = -x^4 - 6x^2$'

plot_lines(title, lines)
```

```python
# y=2/x

# Create a range of x values
x = np.linspace(-10, 10, 100)
# Avoid division by zero
x = x[x != 0]

# Compute the corresponding y values
y = (lambda x: 2 / x)(x)

lines = [{'x': x, 'y': y}]
title = '$y = \\frac{2}{x}$'

plot_lines(title, lines)
```
