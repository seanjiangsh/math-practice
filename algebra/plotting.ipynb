```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def setup_plot():
    """
    Set up the plot with grid, ticks, and limits.
    """    
    plt.axhline(0, color='black', linewidth=1.0)
    plt.axvline(0, color='black', linewidth=1.0)
    plt.grid(which='both', color='gray', linestyle='--', linewidth=0.5)
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='gray')
    plt.xticks(np.arange(-10, 11, 1))
    plt.yticks(np.arange(-10, 11, 1))
    plt.gca().set_xticks(np.arange(-10, 11, 5), minor=False)
    plt.gca().set_yticks(np.arange(-10, 11, 5), minor=False)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
```

```python
from numpy import ndarray
from typing import TypedDict, Optional

class LineDict(TypedDict):
    x: ndarray[float]
    y: ndarray[float]
    color: Optional[str]
    linestyle: Optional[str]
    label: Optional[str]

def plot_lines(lines:list[LineDict], title:str):
    plt.figure(figsize=(6, 6))
    
    for line in lines:
        x = line['x']
        y = line['y']
        color = line.get('color', 'blue')
        linestyle = line.get('linestyle', '-')
        label = line.get('label', None)
        plt.plot(x, y, color=color, linestyle=linestyle, label=label)

    setup_plot()
    plt.xlabel('x')
    plt.ylabel('y', rotation=0)
    plt.title(title)

    # Check if any line has a label and add legend if true
    has_label = any(map(lambda line: isinstance(line.get('label'), str), lines))
    if has_label:
        plt.legend()
    
    plt.show()
```

### Cartesian coordinate of y = -2/3x

```python
def get_slope_intercept_points(slope:float, intercept=0):
    x = np.linspace(-10, 10, 100)
    # Calculate y values based on the equation y = mx + b
    y = slope * x + intercept
    return {'x': x, 'y': y}
```

```python
# y = -2/3x
points = get_slope_intercept_points(-2/3, 0)
title = r'$y = -\frac{2}{3}x$'
plot_lines([points], title)
```

```python
class FillDict(TypedDict):
    x: ndarray[float] # X range for filling
    y1: ndarray[float]  # Y values for the lower boundary
    y2: ndarray[float] # Y values for the upper boundary
    where: ndarray[bool] # Condition for filling
    color: Optional[str] # Color for filling


def plot_inequality(lines:list[LineDict], fills:list[FillDict], title:str):
    """
    Plot an inequality on a number line.
    
    Args:
    lines (list): A list of dictionaries containing the x and y values for the lines.
        x (ndarray): The x values for the line.
        y (ndarray): The y values for the line.
        color (str): The color of the line.
        linestyle (str): The line style for the line, note the "greater/less equal" should be solid, while the "greater/less then" should be dashed.
                         for example, ">=" should be solid, and ">" should be dashed.
        label (str): The label for the line.
    fills (list): A list of dictionaries containing the x range and y values for filling the region.
        x (ndarray): The x range for filling.
        y1 (ndarray): The y values for the lower boundary.
        y2 (ndarray): The y values for the upper boundary.
        where (ndarray): The condition for filling.
        color (str): The color for filling.
    title (str): The title of the plot.
    """
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Plot the lines
    for line in lines:
        x = line['x']
        y = line['y']
        color = line.get('color', 'blue')
        linestyle = line.get('linestyle', '-')
        label = line.get('label', None)
        plt.plot(x, y, color=color, linestyle=linestyle, label=label)

            
    # Fill the region
    for fill in fills:
        x_fill = fill['x']
        y_fill_1 = fill['y1']
        y_fill_2 = fill['y2']
        where = fill['where']
        color = fill.get('color', 'lightblue')
        ax.fill_between(x_fill, y_fill_1, y_fill_2, where, color=color, alpha=0.5)
    
    setup_plot()
    plt.title(title)

    # Check if any line has a label and add legend if true
    has_label = any(map(lambda line: isinstance(line.get('label'), str), lines))
    if has_label:
        plt.legend()

    plt.show()
```

```python
# x > 2
all_range = np.linspace(-10, 10, 100)
x = np.linspace(2, 2, 100)
lines = [{'x': x, 'y': all_range, 'linestyle': '--'}]
fills = [{
    'x': all_range,
    'y1': -10, # lower boundary
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': all_range >= 2
}]
title = '$x > 2$'

plot_inequality(lines, fills, title)
```
