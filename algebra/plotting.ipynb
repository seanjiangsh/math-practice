## Utilities

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from plotting_utils import LineDict, FillDict, LimitDict # Type definitions
from plotting_utils import setup_plot

def plot_lines(title:str, lines:list[LineDict], fills:list[FillDict]=None, limits:LimitDict=None):

    plt.figure(figsize=(6, 6))
    
    for line in lines:
        x = line['x']
        y = line['y']
        line_color = line.get('color', 'blue')
        linestyle = line.get('linestyle', '-')
        label = line.get('label', None)
        plt.plot(x, y, color=line_color, linestyle=linestyle, label=label)

        # Plot points
        points = line.get('points', [])
        for point in points:
            x = point['x']
            y = point['y']
            equal = point['equal']
            point_color = line_color if equal else 'white'
            marker_edge_color = None if equal else line_color
            plt.plot(x, y, 'o', color=point_color, markersize=5, markeredgewidth=1, markeredgecolor=marker_edge_color)
        
    # Plot fills
    if fills is not None:
        for fill in fills:
            x = fill['x']
            y1 = fill['y1']
            y2 = fill['y2']
            where = fill['where']
            fill_color = fill.get('color', 'lightblue')
            plt.fill_between(x, y1, y2, where, color=fill_color, alpha=0.5)


    setup_plot(limits)
    plt.xlabel('x')
    plt.ylabel('y', rotation=0)
    plt.title(title)

    # Check if any line has a label and add legend if true
    has_label = any(map(lambda line: isinstance(line.get('label'), str), lines))
    if has_label:
        plt.legend()
    
    plt.show()
```

## Basic slope-intercept form of line(s)

```python
def get_slope_intercept_points(slope:float, intercept=0):
    x = np.linspace(-10, 10, 100)
    # Calculate y values based on the equation y = mx + b
    y = slope * x + intercept
    return {'x': x, 'y': y}
```

## Piecewise Function

```python
# f(x) = x^2 if x<3
# f(x) = 8 if x=3
# f(x) = 2x + 4 if x>3

x1 = np.linspace(-10, 3, 100)
y1 = x1 ** 2
p1 = {'x': 3, 'y': 3**2, 'equal': False}
label1 = '$f(x) = x^2$ if $x<3$'
line_1 = {'x': x1, 'y': y1, 'label': label1, 'points': [p1]}

x2 = np.linspace(0, 0, 0)
y2 = np.linspace(0, 0, 0)
p2 = {'x': 3, 'y': 8, 'equal': True}
label2 = '$f(x) = 8$ if $x=3$'
line_2 = {'x': x2, 'y': y2, 'label': label2, 'points': [p2], 'color': 'orange'}

x3 = np.linspace(3, 10, 100)
y3 = 2 * x3 + 4
p3 = {'x': 3, 'y': 10, 'equal': False}
label3 = '$f(x) = 2x + 4$ if $x>3$'
line_3 = {'x': x3, 'y': y3, 'label': label3, 'points': [p3], 'color': 'green'}

lines = [line_1, line_2, line_3]
title = 'Piecewise Function'

limits = {'x': (-10, 10), 'y': (0, 20)}

plot_lines(title, lines, limits=limits)
```

## Quadratic Inequalities (Parabola)

```python
# x^2 - 9x + 14 <= 0
# (x - 2)(x - 7) <= 0

discriminant = (-9)**2 - 4 * 1 * 14
print(f'We have 2 real roots (2 zeros) because the discriminant is {discriminant} and it is positive')
vertex_x = -(-9) / (2 * 1)
vertex = (vertex_x, vertex_x**2 - 9 * vertex_x + 14)
print(f'The vertex of the parabola is {vertex}')
print(f'The parabola opens upwards because the coefficient of x^2 is positive')
print('So the parabola will be below the x-axis between the zeros, and the zeros are 2 and 7')
print('Therefore, the solution to the inequality is 2 <= x <= 7')

x = np.linspace(2, 7, 100)
y = (lambda x: x**2 - 9*x + 14)(x)

lines = [{'x': x, 'y': y}]
title = '$x^2 - 9x + 14 \\leq 0$'
limits = {'x': (0, 15), 'y': (-10, 5)}

plot_lines(title, lines, limits=limits)
```

```python
# x^2 + 2x - 3 < 0
# (x + 3)(x - 1) < 0

discriminant = 2**2 - 4*1*(-3)
print(f'We have 2 real roots because the discriminant is {discriminant} and it is positive')
vertex_x = -2 / (2 * 1)
vertex = (vertex_x, vertex_x**2 + 2 * vertex_x - 3)
print(f'The vertex of the parabola is {vertex}')
print(f'The parabola opens upwards because the coefficient of x^2 is positive')
print('So the parabola will be below the x-axis between the zeros, and the zeros are -3 and 1')
print('Therefore, the solution to the inequality is -3 < x < 1')

x = np.linspace(-3, 1, 100)
y = (lambda x: x**2 + 2*x - 3)(x)

lines = [{'x': x, 'y': y}]
title = '$x^2 + 2x - 3 < 0$'
limits = {'x': (-5, 5), 'y': (-5, 5)}

plot_lines(title, lines, limits=limits)
```

```python
# x^2 + 10x + 21 > 0
# (x + 3)(x + 7) > 0

discriminant = 10**2 - 4*1*21
print(f'We have 2 real roots because the discriminant is {discriminant} and it is positive')
vertex_x = -10 / (2 * 1)
vertex = (vertex_x, vertex_x**2 + 10 * vertex_x + 21)
print(f'The vertex of the parabola is {vertex}')
print(f'The parabola opens upwards because the coefficient of x^2 is positive')
print('So the parabola will be above the x-axis of the zeros, and the zeros are -3 and -7')
print('Therefore, the solution to the inequality is x < -7 and x > -3')

x1 = np.linspace(-10, -7, 100)
y1 = (lambda x: x**2 + 10*x + 21)(x1)
line_1 = {'x': x1, 'y': y1, 'label': '$x^2 + 10x + 21 > 0$'}

x2 = np.linspace(-3, 10, 100)
y2 = (lambda x: x**2 + 10*x + 21)(x2)
line_2 = {'x': x2, 'y': y2}

lines = [line_1, line_2]
title = '$x^2 + 10x + 21 > 0$'
limits = {'x': (-15, 5), 'y': (-5, 15)}

plot_lines(title, lines, limits=limits)
```

```python
# -x^2 + x + 12 >= 0
# (x - 4)(x + 3) >= 0

discriminant = 1**2 - 4*(-1)*12
print(f'We have 2 real roots (2 zeros) because the discriminant is {discriminant} and it is positive')
vertex_x = ((-1) / (2 * (-1)))
vertex = (vertex_x, -vertex_x**2 + vertex_x + 12)
print(f'The vertex of the parabola is {vertex}')
print(f'The parabola opens downwards because the coefficient of x^2 is negative')
print('So the parabola will be above the x-axis between the zeros, and the zeros are -3 and 4')
print('Therefore, the solution to the inequality is -3 <= x <= 4')

x = np.linspace(-3, 4, 100)
y = (lambda x: -x**2 + x + 12)(x)
lines = [{'x': x, 'y': y}]

title = '$-x^2 + x + 12 \\geq 0$'
limits = {'x': (-10, 10), 'y': (-5, 15)}

plot_lines(title, lines, limits=limits)
```

## Systems with quadratic inequalities

```python
# y > 2x + 1
# y <= (2x + 1)(x - 3)

x1 = np.linspace(-10, 10, 100)
y1 = 2 * x1 + 1
x2 = np.linspace(-10, 10, 100)
y2 = (2 * x2 + 1) * (x2 - 3)

line_1 = {
    'x': x1,
    'y': y1,
    'linestyle': '--',
    'label': '$y > 2x + 1$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y \\leq (2x + 1)(x - 3)$',
    'color': 'orange'
}

# fill_1 = {
#     'x': x1,
#     'y1': y1,
#     'y2': 10,
#     'where': y1
# }
# fill_2 = {
#     'x': x2,
#     'y1': -10,
#     'y2': y2,
#     'where': y2,
#     'color': '#f2d583'
# }
fill = {
    'x': x1,
    'y1': y1,
    'y2': y2,
    'where': (y1 < y2),
    'color': '#e374b5'
}
# x, y1, y2, where=(y1 < y2)
lines = [line_1, line_2]
# fills = [fill_1, fill_2]
fills = [fill]
title = 'Systems with quadratic inequalities'

plot_lines(title, lines, fills)
```

```python
# y > 5x^2 - 6
# y <= -3x

x1 = np.linspace(-10, 10, 100)
y1 = 5 * x1**2 - 6
x2 = np.linspace(-10, 10, 100)
y2 = -3 * x2

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y > 5x^2 - 6$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y \\leq -3x$',
    'color': 'orange'
}

# fill_1 = {
#     'x': x1,
#     'y1': y1,
#     'y2': 10,
#     'where': y1
# }
# fill_2 = {
#     'x': x2,
#     'y1': -10,
#     'y2': y2,
#     'where': y2,
#     'color': '#f2d583'
# }
fill = {
    'x': x1,
    'y1': y1,
    'y2': y2,
    'where': (y1 < y2),
    'color': '#e374b5'
}

lines = [line_1, line_2]
# fills = [fill_1, fill_2]
fills = [fill]
title = 'Systems with quadratic inequalities'

plot_lines(title, lines, fills)
```

```python
# y >= x^2 - 2
# 5y >= x (y >= 1/5x)

x1 = np.linspace(-10, 10, 100)
y1 = x1**2 - 2
x2 = np.linspace(-10, 10, 100)
y2 = x2 / 5

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y \\geq x^2 - 2$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y \\geq \\frac{1}{5}x$',
    'color': 'orange'
}

fill = {
    'x': x1,
    'y1': np.maximum(y1, y2),
    'y2': 10,
    'where': (y1 <= 10) & (y2 <= 10),
    'color': '#e374b5'
}

lines = [line_1, line_2]
fills = [fill]
title = '$y \\geq x^2 - 2, 5y \\geq x$'

plot_lines(title, lines, fills)
```

```python
# x + y < -1 (y < -x - 1)
# y <= x^2 + 4x + 3

x1 = np.linspace(-10, 10, 100)
y1 = -x1 - 1
x2 = np.linspace(-10, 10, 100)
y2 = x2**2 + 4*x2 + 3

line_1 = {
    'x': x1,
    'y': y1,
    'label': '$y < -x - 1$',
    'linestyle': '--'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y \\leq x^2 + 4x + 3$',
    'color': 'orange'
}

fill = {
    'x': x1,
    'y1': np.minimum(y1, y2),
    'y2': -10,
    'where': (y1 >= -10) & (y2 >= -10),
    'color': '#e374b5'
}

lines = [line_1, line_2]
fills = [fill]
title = '$x + y < -1, y \\leq x^2 + 4x + 3$'

plot_lines(title, lines, fills)
```
