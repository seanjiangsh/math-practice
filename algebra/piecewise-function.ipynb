```python
# The temperature throughout a summer day starts at 65∘ F at 6:00 a.m.. 
# Over the next few hours the temperature rises steadily until it reaches 85∘ F at 1:00 p.m..
# At 1:15 p.m., a rainstorm begins and cools the temperature down to 75∘ F.
# The temperature then steadily decreases until it reaches 70∘ F at 9:00 p.m..

# 65 degrees at 6:00 a.m.
x1 = np.linspace(0, 6, 100)
y1 = np.linspace(0, 65, 100)

# temperature rises to 85 degrees at 1:00 p.m.
x2 = np.linspace(6, 13, 100)
y2 = np.linspace(65, 85, 100)

# rainstorm cools the temperature to 75 degrees at 1:15 p.m.
x3 = np.linspace(13, 13.25, 100)
y3 = np.linspace(85, 75, 100)

# temperature decreases to 70 degrees at 9:00 p.m.
x4 = np.linspace(13.25, 21, 100)
y4 = np.linspace(75, 70, 100)

# temperature stays at 70 degrees until midnight
x5 = np.linspace(21, 24, 100)
y5 = np.linspace(70, 70, 100)

line_1 = {'x': x1, 'y': y1,}
line_2 = {'x': x2, 'y': y2,}
line_3 = {'x': x3, 'y': y3,}
line_4 = {'x': x4, 'y': y4,}
line_5 = {'x': x5, 'y': y5,}

lines = [line_1, line_2, line_3, line_4, line_5]
title = 'Temperature Throughout the Day (x: time, y: temperature in F)'
limits = {'x': (0, 24), 'y': (0, 100)}

plot_lines(title, lines, limits=limits)
```
