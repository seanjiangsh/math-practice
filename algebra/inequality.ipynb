## Practices

```python
# x > 2
all_range = np.linspace(-10, 10, 100)
x = np.linspace(2, 2, 100)
lines = [{'x': x, 'y': all_range, 'linestyle': '--'}]
fills = [{
    'x': all_range,
    'y1': -10, # lower boundary
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': all_range >= 2
}]
title = '$x > 2$'

plot_lines(title, lines, fills)
```

```python
# x <= 5
all_range = np.linspace(-10, 10, 100)
x = np.linspace(5, 5, 100)
lines = [{'x': x, 'y': all_range}]
fills = [{
    'x': all_range,
    'y1': -10, # lower boundary
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': all_range < 5
}]
title = '$x \\leq 5$'

plot_lines(title, lines, fills)
```

```python
# 2 < x <= 5
all_range = np.linspace(-10, 10, 100)
line_1 = {
    'x': np.linspace(2, 2, 100),
    'y': all_range,
    'linestyle': '--',
    'label': '$x > 2$'
}
line_2 = {
    'x': np.linspace(5, 5, 100),
    'y': all_range,
    'label': '$x \\leq 5$',
}
fill = {
    'x': all_range,
    'y1': -10, # lower boundary
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': (all_range > 2) & (all_range <= 5)
}

title = '$2 < x \\leq 5$'

plot_lines(title, [line_1, line_2], [fill])
```

```python
# y >= 1/2x -3
all_range = np.linspace(-10, 10, 100)
x = all_range
y = (lambda x: 0.5 * x - 3)(x)
lines = [{'x': x, 'y': y,}]
fills = [{
    'x': all_range,
    'y1': y, # lower boundary (the line)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': y
}]
title = '$y \\geq \\frac{1}{2}x - 3$'

plot_lines(title, lines, fills)
```

```python
# y > -2x + 1
all_range = np.linspace(-10, 10, 100)
x = all_range
y = (lambda x: -2 * x + 1)(x)
lines = [{'x': x, 'y': y, 'linestyle': '--',}]
fills = [{
    'x': all_range,
    'y1': y, # lower boundary (the line)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': y
}]
title = '$y > -2x + 1$'

plot_lines(title, lines, fills)
```

```python
# y <= 2x + 4
all_range = np.linspace(-10, 10, 100)
x = all_range
y = (lambda x: 2 * x + 4)(x)
lines = [{'x': x, 'y': y}]
fills = [{
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': y, # upper boundary (the line)
    'where': y
}]
title = '$y \\leq 2x + 4$'

plot_lines(title, lines, fills)
```

```python
# x <= 1 or x > 4
all_range = np.linspace(-10, 10, 100)
line_1 = {
    'x': np.linspace(1, 1, 100),
    'y': all_range,
    'label': '$x \\leq 1$',
}
line_2 = {
    'x': np.linspace(4, 4, 100),
    'y': all_range,
    'linestyle': '--',
    'label': '$x > 4$',
}

fill_1 = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': all_range <= 1
}
fill_2 = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': all_range > 4
}

lines = [line_1, line_2]
fills = [fill_1, fill_2]
title = '$x \\leq 1$ or $x > 4$'

plot_lines(title, lines, fills)
```

```python
# all range
all_range = np.linspace(-10, 10, 100)
fill = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': all_range
}
title = 'All Real Numbers'

plot_lines(title, [], [fill])
```

```python
# y < x, y >= -2
all_range = np.linspace(-10, 10, 100)
x = all_range
y1 = x
y2 = np.linspace(-2, -2, 100)
lines = [
    {'x': x, 'y': y1, 'linestyle': '--', 'label': '$y < x$'},
    {'x': x, 'y': y2, 'color': 'orange', 'label': '$y \\geq -2$'}
]
fill_1 = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': y1, # upper boundary (the line)
    'where': y1,
}
fill_2 = {
    'x': all_range,
    'y1': y2, # lower boundary (the line)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': y2,
    'color': '#f2d583'
}
fills = [fill_1, fill_2]
title = '$y < x$, $y \\geq -2$'

plot_lines(title, lines, fills)
```

```python
# y < -2x + 3, y<= 3x - 2
all_range = np.linspace(-10, 10, 100)

x1, y1 = get_slope_intercept_points(-2, 3).values()
x2, y2 = get_slope_intercept_points(3, -2).values()

line_1 = {
    'x': x1,
    'y': y1,
    'linestyle': '--',
    'label': '$y < -2x + 3$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'color': 'orange',
    'label': '$y \\leq 3x - 2$'
}

fill_1 = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': y1, # upper boundary (the line)
    'where': y1
}
fill_2 = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': y2, # upper boundary (the line)
    'where': y2,
    'color': '#f2d583'
}

lines = [line_1, line_2]
fills = [fill_1, fill_2]
title = '$y < -2x + 3$, $y \\leq 3x - 2$'

plot_lines(title, lines, fills)
```

```python
# y = -x + 1, y = x + 1
all_range = np.linspace(-10, 10, 100)
line_1 = {
    'x': all_range,
    'y': -all_range + 1,
    'label': '$y = -x + 1$'
}
line_2 = {
    'x': all_range,
    'y': all_range + 1,
    'label': '$y = x + 1$'
}
title = '$y = -x + 1$, $y = x + 1$'

plot_lines(title, [line_1, line_2])
```

```python
# y > x + 1, y <= -x + 5
x1, y1 = get_slope_intercept_points(1, 1).values()
x2, y2 = get_slope_intercept_points(-1, 5).values()
line_1 = {
    'x': x1,
    'y': y1,
    'linestyle': '--',
    'label': '$y > x + 1$'
}
line_2 = {
    'x': x2,
    'y': y2,
    'label': '$y \\leq -x + 5$',
    'color': 'orange'
}

fill_1 = {
    'x': all_range,
    'y1': y1, # lower boundary (the line)
    'y2': 10, # upper boundary (between -10 and 10, all y values)
    'where': y1
}
fill_2 = {
    'x': all_range,
    'y1': -10, # lower boundary (between -10 and 10, all y values)
    'y2': y2, # upper boundary (the line)
    'where': y2,
    'color': '#f2d583'
}

lines = [line_1, line_2]
fills = [fill_1, fill_2]
title = '$y > x + 1$, $y \\leq -x + 5$'

plot_lines(title, lines, fills)
```
